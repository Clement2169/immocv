```python
from bs4 import BeautifulSoup as bs
import requests 

url_imdb =  "https://assets-datascientest.s3.eu-west-1.amazonaws.com/IMDb.html"

page_imdb = requests.get (url_imdb)

bs_imdb = bs(page_imdb.content,"lxml")

bs_imdb.prettify().splitlines()[0:30]
```

```python
films_imdb = bs_imdb.find_all("td",class_ = "titleColumn")
print (films_imdb[0:30])
```

```python
first_film  = films_imdb[0]
print (first_film)
first_film_title  = first_film.find("a").text
print ("first_film_title :",first_film_title)
```

```python
first_film_rank  = films_imdb[0].text.split() [0].replace(".","")
print ("first_film_rank :",first_film_rank)
```

```python
first_film_year  = films_imdb[0].text.split()[-1].replace("(","").replace(")","")
print ("first_film_year :",first_film_year)
```

```python
notes = bs_imdb.find_all("td",class_ = "ratingColumn imdbRating")
print ("rating of the first film :",float(notes[0].text.strip().replace(',','.')))
```

```python
# Insérez votre code ici
import pandas as pd 

titles = []
ranks = []
years =[]
ratings = []
for film in films_imdb :
    film_rank  = int(film.text.split() [0].replace(".",""))
    film_year  = int(film.text.split()[-1].replace("(","").replace(")",""))
    film_title = film.find("a").text
    
    titles.append(film_title)
    ranks.append(film_rank)
    years.append(film_year)

for  note in notes :
    ratings.append (float(note.text.strip().replace(',','.')))

tables = list(zip(titles,ranks,ratings,years))
df_imdb = pd.DataFrame(tables,columns=["Title","Rank","Rating","Year Out"])
display(df_imdb)
```

```python
url_allocine =  "https://assets-datascientest.s3.eu-west-1.amazonaws.com/AlloCiné.html"

page_allocine = requests.get (url_allocine)

bs_allocine = bs(page_allocine.content,"lxml")

bs_allocine.prettify().splitlines()[0:30]
```

```python
# recover first page films
films_allocine = bs_allocine.find_all('li',class_="mdl")

print ( "number of films of the first page", len(films_allocine))

```

```python
# recover first film
first_film = films_allocine[0]
# extract first film information
first_film_name = first_film.find_next("h2").text.strip()
print ("first_film_name",first_film_name)
```

```python
first_film_ratings = first_film.find_all("span",class_ = "stareval-note")
first_film_press_rating = float (first_film_ratings[0].text.strip().replace(',','.'))
print ("first_film_press_rating",first_film_press_rating)
first_film_spectator_rating = float (first_film_ratings[1].text.strip().replace(',','.'))
print ("first_film_spectator_rating",first_film_spectator_rating)
```

```python
import pandas as pd
film_titles = []
press_ratings = []
spectator_ratings = []
for film in films_allocine :
    film_titles.append(film.find_next("h2").text.strip())
    film_ratings = film.find_all("span",class_ = "stareval-note")

    press_ratings.append (float(film_ratings[0].text.strip().replace(',','.')))
    spectator_ratings.append (float(film_ratings[1].text.strip().replace(',','.')))

tables = list(zip(film_titles,press_ratings,spectator_ratings))
df_allocine = pd.DataFrame(tables,columns=["Title","Press rating","Spectator rating"])
display(df_allocine)
```

```python
df_allocine.Title = df_allocine.Title.str.upper()
df_imdb.Title = df_imdb.Title.str.upper()
display (df_imdb)
display (df_allocine)

imdb_allocine_common = pd.merge(df_imdb, df_allocine, on='Title')

```

```python
sorted_films = imdb_allocine_common.sort_values(by=["Spectator rating"],ascending=False)
display (sorted_films)

print ("best films according to the  spectators are : ", sorted_films.iloc[0].Title," , ",sorted_films.iloc[1].Title)
```
