```python
import numpy as np
import torch

# create tensor
x = torch.ones((3,3), requires_grad=True)

print (x)
```

```python
# create f
def f1 (x) :
    return torch.cos(x*x)

# create g
def g1(x) :
    return torch.pow (torch.tan(x), 1/4)

# retain gradient

f = f1(x)
f.retain_grad()

g= g1(x)
g.retain_grad()

x.grad= None

hr = 1/( 1 + torch.exp(f-g))

s = torch.sum(hr)
print (s)

s.backward()

print(f'after backward Gradient of data for h : {x.grad}')

print(f'after backward Gradient of data for f : {f.grad}')

print(f'after backward Gradient of data for g : {g.grad}')
```
